Börja med att installera och sätta upp ipytest, detta steg kör du bara en gång.

Sedan:
* Skriv din kod (den som skall testas)
* Skriv sedan dina tester (de som testar koden)

De här tre stegen kan du ändra, köra igen och ändra igen och köra igen.
Tänk bara på att köra stegen i ordning.

```python
# Börja med att installera och sätta upp ipytest (detta gör du en gång)
!pip install pytest ipytest

import ipytest
ipytest.autoconfig()
```

```python
# Här skriver du din kod (ditt python-skript)
def add(a, b):
    return a + b

def divide(a, b):
    if b == 0:
        raise ValueError("Division by zero")
    return a / b
```

```python
# Här skriver du dina pytest-tester
# Tänk på att de skall börja med test_

def test_add():
    assert add(2, 3) == 5
    assert add(-1, 1) == 0

def test_divide():
    assert divide(10, 2) == 5
    with pytest.raises(ValueError):
        divide(10, 0)
```

```python
# Kör testerna
ipytest.run("-vv")   # -vv ger mer detaljerad output
```
